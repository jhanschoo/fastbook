We illustrate how the quadratics look and the penalty they impose on larger-valued weights.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)
a_s = [1, 2, 5, 10, 50] 
ys = [a * x**2 for a in a_s]
_, ax = plt.subplots(figsize=(8, 6))
for a, y in zip(a_s, ys): ax.plot(x, y, label=f'a={a}')
ax.set_ylim([0, 5])
ax.legend();
```
